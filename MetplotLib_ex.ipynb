```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```
x=np.arange(10,110,10)
print(x)
y=np.arange(15,165,15)
print(y)
y1=np.arange(10,65,5)
print(y1)
#plt.plot(x,y,marker='<',color='#8899ff',label=3030,mec='g')
plt.plot(x,y,marker='<',color='#8899ff',label=6080,mec='r')
plt.xlabel('House area')
plt.ylabel('price in 100')
plt.title('House Price graph')
plt.legend(ncol=2,loc='best')
plt.show()
```

```
x=np.arange(-3,3,0.005)
print(x)
y=np.sin(x)
y1=np.cos(x)
#plt.plot(x,y,'r*',label='sine')
plt.plot(x,y1,'b+',label='cos')
plt.xlabel('Degree')
plt.ylabel('cos value')
plt.title('cos graph')
plt.legend()
plt.show()
```

```
fig,ax=plt.subplots(2,2,figsize=(5,5))
x=np.arange(1,5)
ax[0][0].plot(x,x*x,color='r')
ax[0][0].set_title('square')
ax[0][1].plot(x,np.exp(x),color='yellow')
ax[0][1].set_title('square root')
ax[1][0].plot(x,np.sqrt(x),color='black')
ax[1][0].set_title('exp')
ax[1][1].plot(x,np.log10(x),color='k')
ax[1][1].set_title('log')
plt.show()
```

```
fig,ax=plt.subplots(2,2,figsize=(7,7))
x=np.arange(1,5)
ax[0][0].plot(x,x*x,color='r')
ax[0][0].grid(True)
ax[0][0].set_title('square')
ax[0][1].plot(x,np.exp(x),color='yellow')
ax[0][1].grid(color='g',ls='-',lw=0.35)
ax[0][1].set_title('square root')
ax[1][0].plot(x,np.sqrt(x),color='black')
ax[1][0].grid(color='b',ls='-',lw=0.35)
ax[1][0].set_title('exp')
ax[1][1].plot(x,np.log10(x),color='k')
ax[1][1].grid(color='y',ls='-',lw=0.35)
ax[1][1].set_title('log')

plt.show()
```

```
data=[[30,25,50,20],[10,23,51,17],[35,22,45,19]]
x=np.arange(4)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(x,data[0],color='b',width=0.35)
ax.bar(x+0.25,data[1],color='g',width=0.25)
ax.bar(x+0.5,data[2],color='r',width=0.25)
```

```
water=[10,20,50,60,40]
oil=[90,80,50,60,70]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
x=np.arange(5)
ax.bar(x,water,width=0.35,color='lightblue')
ax.bar(x,oil,width=0.35,bottom=water,color='y')
ax.set_xlabel('Indicater')
ax.set_ylabel('Value')
ax.legend(labels=['water','oil'])
ax.set_title('Water oil Relation')
```
